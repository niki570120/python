# 讀取 scipy 內建影像，並轉為灰階

```python
face = scipy.misc.face(gray=True)
plt.imshow(face, cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```

## 圓以外設成黑色(0)，白色(255)

```python
lx, ly = face.shape
# 取得影像 X、Y 座標
X, Y = np.ogrid[0:lx, 0:ly]

# 圓以外設成黑色(0)
mask = (X - lx/2)**2 + (Y - ly/2)**2 > lx*ly/4
face[mask] = 255
```

```python
X.shape, Y.shape
```

```python
plt.imshow(face, cmap=plt.cm.gray)
plt.axis('off')
```

# 縮小圓

```python
# 縮小圓, lx*ly/4 ==> lx*ly/6
# 圓以外設成黑色(0)
mask = (X - lx/2)**2 + (Y - ly/2)**2 > lx*ly/6 
face[mask] = 255

plt.imshow(face, cmap=plt.cm.gray)
plt.axis('off')
```

# 讀取影像檔，並轉為灰階

```python
import matplotlib.image as mpimg

def rgb2gray(rgb):
    print(rgb.shape)
    # gray = 0.2989 R + 0.5870 G + 0.1140 B 
    return np.dot(rgb, [0.2989, 0.5870, 0.1140])

img = mpimg.imread('./lena.jpg')     
gray = rgb2gray(img)    
plt.imshow(gray, cmap='gray')
plt.axis('off')
plt.show()
```

# 圓形特效

```python
img = mpimg.imread('./lena.jpg')   

lx, ly, lz = img.shape
# 取得影像 X、Y 座標
X, Y, Z = np.ogrid[0:lx, 0:ly, 0:lz]
X.shape, Y.shape, Z.shape
```

```python
type(img), img.shape
```

```python
img.flags
```

## RGB 分別處理

```python
img2 = np.array(img.copy())
mask = (X - lx/2)**2 + (Y - ly/2)**2 > lx*ly/6 
R = img2[:,:,0:1]
R[mask] = 0
# R.shape

G = img2[:,:,1:2]
G[mask] = 0

B = img2[:,:,2:]
B[mask] = 0

mask_rgb = np.concatenate((R,G,B), axis=2) 
mask_rgb.shape
```

```python
plt.imshow(mask_rgb) 
plt.axis('off')
plt.show()
```

## 逐點處理(pixel-wised)

```python
# 圓以外設成黑色(0)
# mask = (X - lx/2)**2 + (Y - ly/2)**2 > lx*ly/6 
img2 = img.copy()
for i in range(lx):
    for j in range(ly):
        if (i - lx/2)**2 + (j - ly/2)**2 > lx*ly/6:
            img2[i,j,:] = 255

plt.imshow(img2) 
plt.axis('off')
plt.show()
```

# 黑白、彩色混合

```python
gray = rgb2gray(img)    
gray.shape
```

```python
gray = gray.reshape(*gray.shape,1)
gray.shape
```

```python
gray_rgb = np.concatenate((gray, gray, gray), axis=2) / 255
gray_rgb.shape
```

```python
plt.imshow(gray_rgb)
plt.axis('off')
plt.show()
```
